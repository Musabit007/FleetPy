### This code Generate a Corresponding Node index for OSMID on the edge Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load the node data from OpenStreetMap

```python
# Load the node data from OpenStreetMap
nodes = pd.read_csv('Kentucky_nodes.csv')
nodes
```

## Create Node Dictionary

```python
node_lookup = dict(zip(nodes['osm_id'], nodes['node_index']))
node_lookup
```

```python
node_lookup[6009457934]
```

# Load KY edges from OpenStreetMap

```python
# Load KY edges from OpenStreetMap
edges = pd.read_csv('ky_edges_new.csv')
edges
```

```python
edges = pd.DataFrame(edges)

edges
```

```python
# Create New Columns 
```

```python
# Creates a column with NaN values
edges['from_node'] = None  
```

```python
# Creates a column with NaN values
edges['to_node'] = None  
```

```python
# Rename columns appropriately 
edges = edges.rename(columns={'u': 'from_osm', 'v': 'to_osm', 'length': 'distance', 'osmid': 'source_edge_id', 'highway': 'road_type'})
```

```python
edges
```

```python
# Select only the relevant columns
edges = edges[['from_node', 'to_node', 'from_osm', 'to_osm', 'distance', 'travel_time', 'source_edge_id', 'road_type', 'geometry']]
```

```python
edges
```

```python
# Apply the node lookup
edges['from_node'] = edges['from_osm'].apply(lambda x : node_lookup[x])
```

```python
# Apply the node lookup
edges['to_node'] = edges['to_osm'].apply(lambda x : node_lookup[x])
```

```python
edges
```

```python
# Check for NaN value in the dataset

edges.isna().any()
```

```python
# Save edges as a csv
edges.to_csv('Kentucky_edges_enhanced.csv', index=False)
```
